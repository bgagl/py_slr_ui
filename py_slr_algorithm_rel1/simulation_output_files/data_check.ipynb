```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()

T = 100
a = np.pi/5

A = np.array([[np.cos(a), -np.sin(a)],
			  [np.sin(a), np.cos(a)]])
C = np.array([[1., -1.], [-1., 1.]])

Q = np.identity(A.shape[0])
R = np.array([[1, 4], [2, 4]])

x = np.zeros((T, A.shape[0]))
y = np.zeros((T, C.shape[0]))
x[0] = [1., 0]
y[0] = C @ x[0] + np.random.randn(1)

mu  = np.zeros(x.shape)
muh = np.zeros(x.shape)
si  = np.zeros((T, A.shape[0], A.shape[1]))
sih = np.zeros((T, A.shape[0], A.shape[1]))

inn = np.zeros(y.shape)
ico = np.zeros((T, R.shape[0], R.shape[1]))
kal = np.zeros((T, C.shape[1], C.shape[0]))
res = np.zeros(y.shape)

for t in range(1,T):

	# System
	ep = np.random.multivariate_normal(np.zeros(x.shape[1]), Q)
	nu = np.random.multivariate_normal(np.zeros(y.shape[1]), R)
	x[t] = A @ x[t-1] + ep #ep ... process noise
	y[t] = C @ x[t] + nu #nu ... observation noise

	# Prior
	muh[t] = A @ mu[t-1]
	sih[t] = A @ si[t-1] @ A.T + np.identity(A.shape[0])

	# Posterior
	inn[t] = y[t] - C @ muh[t]
	ico[t] = C @ sih[t] @ C.T + R
	kal[t] = sih[t] @ C.T @ np.linalg.inv(ico[t])

	mu[t] = muh[t] + kal[t] @ inn[t]
	si[t] = (np.identity(y.shape[1]) - kal[t] @ C) @ sih[t]


plt.plot(np.arange(T), x[:,0], 'b', label='latent (dim1)')
plt.plot(np.arange(T), mu[:,0], 'b--', label='posterior mean (dim1)')
plt.plot(np.arange(T), x[:,0] - mu[:,0], 'b:', label='prediction error (dim1)')

plt.plot(np.arange(T), x[:,1], 'r', label='latent (dim1)')
plt.plot(np.arange(T), mu[:,1], 'r--', label='posterior mean (dim2)')
plt.plot(np.arange(T), x[:,1] - mu[:,1], 'r:', label='prediction error (dim2)')

plt.xlabel('time')
plt.legend()

plt.show()

sns.jointplot(x=x[:,0]-mu[:,0], y=x[:,1]-mu[:,1], kind="kde", fill=True, space=0)
plt.xlabel('prediction error (dim1)')
plt.ylabel('prediction error (dim2)')
plt.show()
```
